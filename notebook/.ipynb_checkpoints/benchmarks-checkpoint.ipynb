```python
import json

f1="/Users/guillaumelaunay/work/tmp/ccmap2/benchmark_lzmap.log"
zmapBench = {}

with open(f1, "r") as fp:
    for l in fp:
        lot, cpu, time = l.split()
        if not lot in zmapBench:
            zmapBench[lot] = []
        zmapBench[lot].append([int(cpu), float(time)])
```
